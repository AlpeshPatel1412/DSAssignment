Q1. Explain GET and POST methods.

GET - It is the common method used to send data in the unencrypted form to the server. By default, the requests are handled by GET method.
The data passed by GET method are visible in URL and hence it is not secure

POST - It is used to send the form data to the server. The server does not cache the data transmitted using the post method. The data passed to server are not visible to users


Q2. Why is request used in Flask?

A2. Request object contains all the data that is sent from the client to the server. we can retrieve the data at the server side using the HTTP  methods from the request object. 

Below are the important attributes of the request object.

Form - It is the dictionary object which contains the key-value pair of form parameters and their values.

args - It is parsed from the URL. It is the part of the URL which is specified in the URL after question mark (?).

Cookies - It is the dictionary object containing cookie names and the values. It is saved at the client-side to track the user session.

files -	It contains the data related to the uploaded file.

method - It is the current request method (get or post).

Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask, jsonify, request

app = Flask(__name__)


@app.route("/postname_demo", methods=['POST'])
def convert_temp():
    if (request.method == 'POST'):
        ops = request.json['operation']
        input_temp = float(request.json['temp'])

        if (ops == 'Fahrenheit_to_Celsius'):
            output_temp = (input_temp-32)*5/9
            result = str(input_temp) + " Fahrenheit = " + str(output_temp) + " Celsius"
        elif (ops == 'Celsius_to_Fahrenheit'):
            output_temp = (input_temp * 9/5) + 32
            result = str(input_temp) + " Celsius = " + str(output_temp) + " Fahrenheit"
        else:
            result = "Invalid Operation"
        
        return jsonify(result)
   


if __name__=="__main__":
    app.run(host="0.0.0.0")
```
